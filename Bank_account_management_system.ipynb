```python
"""Below is the attached output"""

import random
import datetime    # importing the necessasry modules
credentials = {}   # Creating empty dictionaries
users_db = {}
passbook = {}
nested_trans = {}

print("~"*10, "Welcome to XYZ Bank", "~"*10)
print("\nHow can we Help you Today ?")

while True : 
    print("""
1. Open an Account.
2. View Accounts Details.
3. Perform Transactions (Withdraw/Deposit/Transfer).
4. View Transaction History.
5. Exit""")

    try :                            # Try except for exception handling
        choice = int(input("\nEnter your choice between 1 to 5 : "))
    except ValueError :                                                       
        print("Invalid choice! Please choose a value between 1 to 5")
        continue
    
    if choice == 1 : 
        fname = input("Enter First Name : ")
        lname = input("Enter Last Name : ")
        pancard = input("Enter Pancard number : ")

        if pancard in users_db.keys() :
            print("You already have an account with the bank .\nTry Again!!")
            break

        account_number = "99990000" #The initial 8 digits of account number will be constant to represent Bank branch
        for i in range(1,5):
            a = random.randint(0,9)
            account_number +=str(a)

        account_type = input("Which type of account you want to open(Current/Savings) : ")
        try :
            balance = int(input("How much amount you wish to deposite ? "))
        except :
            print("Invalid amount. Please enter a numeric value.")
            continue
        
        username = fname        #username generation with 2 random numeric values at the end of fname
        for i in range(2):
            a = random.randint(0,9)
            username += str(a)

        password = ""
        for i in range(1,5):
            a = random.randint(0,9)
            password += str(a)

        print("Username Assigned : ", username)
        print("Password Assigned : ", password)
        print("Account number : ", account_number)
        print("Account_type : ", account_type)
        print("Initial Account Balance : ", balance,"\n")

        credentials[username] = password      #Assigning key-values
        users_db[pancard] = [fname, lname, account_number, account_type, balance]    #Assigning key-values

    elif choice == 2 :
        un = input("Enter username : ")
        pwd = input("Enter password : ")
        pan = input("Enter PAN no. : ")
        if un in credentials.keys() and pwd == credentials[un] and pan in users_db.keys() :
            print("Your Account details are")
            print("Account Holder name : ", users_db[pan][0],users_db[pan][1])
            print("Account number : ", users_db[pan][2])
            print("Account type : ", users_db[pan][-2])
            print("Total Account Balance : ",users_db[pan][-1])

        else : 
            print("This account does not exists, try again!!")

    elif choice == 3 :
        un = input("Enter username : ")
        pwd = input("Enter password : ")
        pan = input("Enter PAN no. : ")
        if un in credentials.keys() and pwd == credentials[un] and pan in users_db.keys() :
            print("""
    1. Deposit
    2. Withdraw
    3. Transfer""")
            try:
                t_type = input("Enter transaction option: ")
            except ValueError:
                print("Invalid input. Please enter a numeric value.")
                continue
    
            if t_type == "Deposit" :
                try:
                    transfer_amount = float(input("Enter amount you wish to deposit : "))
                    if transfer_amount <= 0 :
                        print("Deposit amount must be greater than 0")
                        continue
                    users_db[pancard][-1] += transfer_amount
                    timestamp = datetime.datetime.now()
                    print("Your current balance is : ", users_db[pancard][-1])
                except ValueError:
                    print("Invalid value! Please enter a numeric value")
    
            elif t_type == "Withdraw" :
                try:
                    withdraw_amount = float(input("Enter amount to withdraw: "))
                    if withdraw_amount <= 0:
                        print("Withdrawal amount must be greater than zero.")
                        continue
                    if withdraw_amount > users_db[pancard][-1] : 
                        print("Insufficient Balance")
                    else : 
                        users_db[pancard][-1] -= withdraw_amount
                        timestamp = datetime.datetime.now()
                        print("Remaining balance is : ", users_db[pancard][-1])
                except ValueError:
                    print("Invalid amount. Please enter a numeric value.")
    
            elif t_type == "Transfer":
                r_pancard = input("Enter receiver's pancard number you wish to transfer money to : ")
                transfer_amount = float(input("Enter amount you wish to transfer : "))
                users_db[pancard][-1] -= transfer_amount
                timestamp = datetime.datetime.now()
                print("Remaining Balance in account is  : ", users_db[pancard][-1])
    
                # If receiver's pancard exists in the same XYZ bank.
                if r_pancard in users_db.keys():
                    users_db[pan][-1] += transfer_amount
    
            nested_trans[timestamp] = [t_type, transfer_amount, users_db[pancard][-1]]
            passbook[pancard] = nested_trans
    

    elif choice == 4 :
        un = input("Enter username : ")
        pwd = input("Enter password : ")
        pan = input("Enter PAN no. : ")
        if un in credentials.keys() and pwd == credentials[un] and pan in users_db.keys() :
            print("          Timestamp           Type    Amount     Balance")
            for i,j in passbook.items():
                for k,l in j.items():
                    print(k,l,end = "\n")
                print()

    elif choice == 5:  # Exit
        print("Thank you for using XYZ Bank. Have a great day!")
        break

    else:
        print("Invalid choice. Please enter a number between 1 and 5.")




"""
            OUTPUT
~~~~~~~~~~ Welcome to XYZ Bank ~~~~~~~~~~

How can we Help you Today ?

1. Open an Account.
2. View Accounts Details.
3. Perform Transactions (Withdraw/Deposit/Transfer).
4. View Transaction History.
5. Exit

Enter your choice between 1 to 5 :  1
Enter First Name :  shivani
Enter Last Name :  nagar
Enter Pancard number :  aaaaa12
Which type of account you want to open(Current/Savings) :  current
How much amount you wish to deposite ?  1000
Username Assigned :  shivani80
Password Assigned :  3191
Account number :  999900005757
Account_type :  current
Initial Account Balance :  1000 


1. Open an Account.
2. View Accounts Details.
3. Perform Transactions (Withdraw/Deposit/Transfer).
4. View Transaction History.
5. Exit

Enter your choice between 1 to 5 :  3
Enter username :  shivani80
Enter password :  3191
Enter PAN no. :  aaaaa12

    1. Deposit
    2. Withdraw
    3. Transfer
Enter transaction option:  Deposit
Enter amount you wish to deposit :  2000
Your current balance is :  3000.0

1. Open an Account.
2. View Accounts Details.
3. Perform Transactions (Withdraw/Deposit/Transfer).
4. View Transaction History.
5. Exit

Enter your choice between 1 to 5 :  3
Enter username :  shivani80
Enter password :  3191
Enter PAN no. :  aaaaa12

    1. Deposit
    2. Withdraw
    3. Transfer
Enter transaction option:  Withdraw
Enter amount to withdraw:  500
Remaining balance is :  2500.0

1. Open an Account.
2. View Accounts Details.
3. Perform Transactions (Withdraw/Deposit/Transfer).
4. View Transaction History.
5. Exit

Enter your choice between 1 to 5 :  4
Enter username :  shivani80
Enter password :  3191
Enter PAN no. :  aaaaa12
          Timestamp           Type    Amount     Balance
2024-11-17 14:09:10.784133 ['Deposit', 2000.0, 3000.0]
2024-11-17 14:09:44.748014 ['Withdraw', 2000.0, 2500.0]


1. Open an Account.
2. View Accounts Details.
3. Perform Transactions (Withdraw/Deposit/Transfer).
4. View Transaction History.
5. Exit

Enter your choice between 1 to 5 :  5
Thank you for using XYZ Bank. Have a great day!
"""
                
                    
        
                
                
            


            
        
            

```
